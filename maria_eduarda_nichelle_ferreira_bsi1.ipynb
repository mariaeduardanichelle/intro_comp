```python
def input_base_choice():
  """
  ¬ Apresenta um menu com as opções de bases de entrada a serem escolhidas.
  :return: O valor numérico da base.
  """
  base = int(input('''
  Escolha a base de entrada:
  [ 1 ] Converter de binário
  [ 2 ] Converter de Octal
  [ 3 ] Converter de Decimal
  [ 4 ] Converter de Hexadecimal
  Sua opção: '''))
  if base < 1 or base > 4:
     raise ValueError('Base inválida!')
  else:
     match base:
      case 1:
        return 2
      case 2:
        return 8
      case 3:
        return 10
      case 4:
        return 16

def output_base_choice():
  """
  ¬ Apresenta um menu com as opções de bases de saída a serem escolhidas.
  :return: O valor numérico da base.
  """
  base = int(input('''
  Escolha a base de saída:
  [ 1 ] Converter para binário
  [ 2 ] Converter para Octal
  [ 3 ] Converter para Decimal
  [ 4 ] Converter para Hexadecimal
  Sua opção: '''))
  if base < 1 or base > 4:
     raise ValueError('Base inválida!')
  else:
     match base:
      case 1:
        return 2
      case 2:
        return 8
      case 3:
        return 10
      case 4:
        return 16

def base_compatibility(num, base):
  """
  ¬ Verifica a compatibilidade do número digitado com a base de entrada.
  :param num: O número a ser convertido.
  :param base: A base de entrada do número.
  :return: Lista com os dígitos do número, caso base e número sejam compatíveis.
  """
  separate_digits = list()
  num = str(num).upper()
  for d in num:
    if d == 'A':
      d = '10'
    elif d == 'B':
      d = '11'
    elif d == 'C':
      d = '12'
    elif d == 'D':
      d = '13'
    elif d == 'E':
      d = '14'
    elif d == 'F':
      d = '15'
    separate_digits.append(int(d))
    match base:
      case 2:
        if int(d) > 1 or int(d) < 0:
          raise ValueError('A base não é compatível com o número!')
      case 8:
        if int(d) > 7 or int(d) < 0:
          raise ValueError('A base não é compatível com o número!')
      case 10:
        if int(d) > 9 or int(d) < 0:
          raise ValueError('A base não é compatível com o número!')
      case 16:
        if int(d) > 15 or int(d) < 0:
          raise ValueError('A base não é compatível com o número!')
  return separate_digits

def current_base_to_dec(num, base):
    """
    ¬ Converte um número da base de entrada para a base decimal.
    :param num: O número a ser convertido.
    :param base: A base de entrada do número.
    :return: O número convertido da base de entrada para decimal.
    """
    digits = base_compatibility(num, base)
    new_num = list()
    decimal_num = 0
    digits.reverse()
    for i, n in enumerate(digits):
      new_num.append(n * base ** i)
    for n in new_num:
      decimal_num += n
    return decimal_num

def decimal_to_base(num, base):
  """
  ¬ Converte um número da base de decimal para a base de saída.
  :param num: O número a ser convertido.
  :param base: A base de saída do número.
  :return: O número convertido da base decimal para a base de saída.
  """
  new_num = list()
  result = ''
  while True:
    remainder = str(num % base)
    if remainder == '10':
      remainder = 'A'
    elif remainder == '11':
      remainder = 'B'
    elif remainder == '12':
      remainder = 'C'
    elif remainder == '13':
      remainder = 'D'
    elif remainder == '14':
      remainder = 'E'
    elif remainder == '15':
      remainder = 'F'
    new_num.insert(0, remainder)
    num = num // base
    if num == 0:
      break
  for n in new_num:
    result += n
  return result

def base_converter(num, input_base, output_base):
  """
  ¬ Converte um número da base de entrada para a base de saída.
  :param num: O número a ser convertido.
  :param input_base: A base de entrada do número.
  :param output_base: A base de saída do número.
  :return: O número convertido da base entrada para a base de saída.
  """
  num = current_base_to_dec(num, input_base)
  num = decimal_to_base(num, output_base)
  return num

num = input('Digite o número: ').upper()
input_base = input_base_choice()
match input_base:
  case 2:
    input_base_name = 'Binária'
  case 8:
    input_base_name = 'Octal'
  case 10:
    input_base_name = 'Decimal'
  case 16:
    input_base_name = 'Hexadecimal'
output_base = output_base_choice()
match output_base:
  case 2:
    output_base_name = 'Binária'
  case 8:
    output_base_name = 'Octal'
  case 10:
    output_base_name = 'Decimal'
  case 16:
    output_base_name = 'Hexadecimal'
print(f'{num} convertido da base {input_base_name} para a base {output_base_name} é {base_converter(num, input_base, output_base)}')
```
